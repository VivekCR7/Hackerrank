```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
def insert_in_DL(head,data):
    newnode = Node(data)
    #if not head
    if not head:
        head = newnode
    
    #place before head
    elif data < head.data:
        newnode.next = head
        head.prev = newnode
        head = newnode
    
    #place at seprate pos
    else:
        #traverse to the curr
        curr = head
        while curr.next != None and curr.data < data:
            curr = curr.next
            
        #place at the end
        if curr.next == None and curr.data < data:
            curr.next = newnode
            newnode.prev = curr
        
        else:
            #separte pos
            previous = curr.prev
            previous.next = newnode
            newnode.prev = previous

            newnode.next =curr
            curr.prev = newnode
    return head
        
    
             
def takeinput(): 
    inputlist = [int(ele) for ele in input().split()]
    head = tail = None
    for curr_data in inputlist:
        if curr_data == -1:
            break
        
        newnode = Node(curr_data)
        
        if head is None:
            head = tail = newnode
        else:
            tail.next = newnode
            tail = newnode
            tail.prev = head
            
    return head

def printLL(head):
    while head is not None:
        if head.next:
            print(str(head.data)+"<->",end="")
        else:
            print(str(head.data),end=".")
        head = head.next
        
    return

head = takeinput()
data = int(input())
head = insert_in_DL(head,data)
printLL(head)


            
```
