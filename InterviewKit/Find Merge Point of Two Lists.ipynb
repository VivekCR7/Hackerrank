```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def MergePoint(head1,head2):
    
    def getCount(head):
        n = 0
        while head is not None:
            n+=1
            head = head.next
        return n
    
    def getNode(d,head1,head2):
        #traverse to d
        for i in range(d):
            head1 = head.next
            
        #check common node
        while head1 and head2:
            if head1 == head2:
                return head1.data
            else:
                head1 = head1.next
                head2 = head2.next
    
    #getcount
    c1 = getCount(head1)
    c2 = getCount(head2)
    
    #check the difference
    if c1>c2:
        return getNode(c1-c2,head1,head2)
    else:
        return getNode(c2-c1,head2,head1)
        
def Takeinput():
    inputlist = [int(ele) for ele in input().split()]
    head = tail = None
    
    for curr_data in inputlist:
        if curr_data == -1:
            break
        
        newnode = Node(curr_data)
        
        if head is None:
            head = tail = newnode
        else:
            tail.next = newnode
            tail = newnode
            
    return head

def printLL(head):
    while head is not None:
        if head.next:
            print(str(head.data)+"->",end="")
        else:
            print(str(head.data),end=".")
        head =head.next
    return

def length(head):
    if head is None:
        return 0
    return 1 + length(head.next)

index = int(input())
head1_count=int(input())
head1 = Takeinput()
head2_count = int(input())
head2 = Takeinput()

ptr1 = head1
ptr2 = head2

for i in range(head1_count):
    if i<index:
        ptr1 = ptr1.next
for i in range(head2_count):
    if i != head2_count-1:
        ptr2 = ptr2.next
        
ptr2.next = ptr1

ans = MergePoint(head1,head2)
print(ans)
```
